```python
# Task 1: String Formatting
#Step 1: Variables
resistance = 10
current = 2.5324
voltage = resistance * current
```

```python
# Step 2: Old Style Formatting (%)
print("The resistor value is %d Ω" % resistance)
print("The current is %.2f A" % current)
print("The voltage across the resistor is %.2f V" % voltage)
```

```python
# Step 3: .format() Method
print("Resistor: {} Ω, Current: {} A, Voltage: {} V"
      .format(resistance, current, voltage))

print("Voltage across {0} Ω resistor with {1} A current is {2} V"
      .format(resistance, current, voltage))
```

```python
# Step 4: f-Strings
print(f"Resistor: {resistance} Ω, Current: {current:.2f} A, Voltage: {voltage:.2f} V")
```
