```python
import math

R = 50
L = 0.1
C = 100e-6
f = 60

w = 2 * math.pi * f

Z_R = complex(R, 0)
Z_L = complex(0, w * L)
Z_C = complex(0, -1 / (w * C))

print("Resistor Impedance:", Z_R)
print("Inductor Impedance:", Z_L)
print("Capacitor Impedance:", Z_C)
```

```python
import cmath

Z_total = Z_R + Z_L + Z_C

print("Total Impedance:", Z_total)
print("Real Part:", Z_total.real)
print("Imaginary Part:", Z_total.imag)
print("Magnitude:", abs(Z_total))
print("Phase Angle:", cmath.phase(Z_total))
```
