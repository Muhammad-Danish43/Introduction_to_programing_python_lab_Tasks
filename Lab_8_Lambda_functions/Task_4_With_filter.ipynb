```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)

words = ["cat", "house", "pen", "apple"]
long_words = list(filter(lambda x: len(x) > 3, words))
print(long_words)

mixed = [1, "hi", 3.5, "hello", True]
strings_only = list(filter(lambda x: type(x) == str, mixed))
print(strings_only)
```

**What we did:**

filter() keeps items that return True.

Lambda was used to write conditions.

We filtered numbers, strings, and mixed data.
