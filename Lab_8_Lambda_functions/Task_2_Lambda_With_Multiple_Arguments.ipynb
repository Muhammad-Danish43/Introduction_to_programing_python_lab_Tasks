```python
# Lambda with two arguments
add = lambda a, b: a + b
print(add(3, 4))

# Lambda with three arguments
multiply = lambda a, b, c: a * b * c
print(multiply(2, 3, 4))

# Lambda with keyword-only argument
power = lambda x, *, y: x ** y
print(power(2, y=3))
```

**What we did:**

We created lambdas that take more than one argument.

Lambda can take any number of arguments.

Keyword-only arguments also work in lambda.
