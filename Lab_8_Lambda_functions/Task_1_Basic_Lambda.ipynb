```python
# Lambda to add 2 to a number
add_two = lambda x: x + 2

print(add_two(3))
print(add_two(10))
```

**What we did**:

We created a lambda function that adds 2 to a number.

We tested it with two different inputs.

A normal function would need def, but lambda is short and one-line only.

**Normal function version:**

```python
def add_two(x):
    return x + 2
```
