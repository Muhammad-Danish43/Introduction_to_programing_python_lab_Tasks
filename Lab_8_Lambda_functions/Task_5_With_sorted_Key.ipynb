```python
words = ["banana", "apple", "kiwi", "grape"]
sorted_words = sorted(words, key=lambda x: len(x))
print(sorted_words)

marks = [("Ali", 80), ("Sara", 90), ("Ahmed", 70)]
sorted_marks = sorted(marks, key=lambda x: x[1])
print(sorted_marks)

scores = {"Ali": 85, "Sara": 92, "Ahmed": 78}
sorted_scores = sorted(scores.items(), key=lambda x: x[1])
print(sorted_scores)
```

**What we did:**

sorted() can use a key function.

Lambda tells Python what to sort by.

We sorted lists and dictionary items easily.
