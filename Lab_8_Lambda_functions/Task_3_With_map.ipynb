```python
numbers = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, numbers))
print(result)

names = ["ali", "sara", "ahmed"]
upper_names = list(map(lambda x: x.upper(), names))
print(upper_names)

points = [(1, 2), (3, 4), (5, 6)]
sums = list(map(lambda x: x[0] + x[1], points))
print(sums)
```

**What we did:**

map() applies a function to each item in a list.

Lambda made the function short and simple.

We used map with numbers, strings, and tuples.
