```python
temps = [23, 25, 19, 30, 21, 22]

print(temps[0])     
print(temps[-1])    
print(temps[-2])    

print(temps[1:5])

print(temps[-2])
```

 **What we did:**

We used indexing to get single values.

We used negative indexing to access values from the end.

We used slicing to get a part of the list.
