```python
products = ["pencil", "notebook", "eraser", "marker"]

products.append("sharpener")
products.append("pen")

products.insert(2, "scale")

print(products)
```

**What we did**:

We created a list of product names.

We used .append() to add new products at the end of the list.

We used .insert() to add a product at a specific position.

A list is suitable here because products can be added or removed easily.
