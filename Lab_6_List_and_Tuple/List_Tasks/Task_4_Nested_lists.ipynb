```python
students = [
    ["Ali", 101, [80, 90, 85]],
    ["Sara", 102, [75, 88, 92]],
    ["Ahmed", 103, [70, 78, 80]]
]

# Change second grade of first student
students[0][2][1] = 95

# Add a fourth grade
students[1][2].append(89)

print(students)
```

**What we did:**

We created a nested list (list inside a list).

Each student has a name, roll number, and grades list.

We accessed and changed a specific grade using indexing.

We added a new grade using .append().
