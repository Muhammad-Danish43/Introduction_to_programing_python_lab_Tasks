```python
weekdays = ("Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun")

print(weekdays[0])
print(weekdays[-1])

print(weekdays[2:5])

print(weekdays[-2])
```

**What we did:**

We accessed tuple elements using indexing.

We used slicing to get specific days.

We used negative indexing to get values from the end.
