```python
device = ("device01", "192.168.1.10", 8080, True)

print(device[0])
print(device[1])
print(device[2])
print(device[3])

device_list = list(device)
device_list[2] = 9090
device = tuple(device_list)

print(device)
```

**What we did**:

We accessed tuple values using indexing.

We tried to change a value but got an error because tuples are immutable.

We converted the tuple to a list, modified it, and converted it back.
